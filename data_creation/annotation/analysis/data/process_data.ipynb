```python
import pickle as pkl
import json
import csv

csv.field_size_limit(1131072)
```

```python
output = {}
with open('results.csv', 'r') as f:
    j = 0
    reader = csv.DictReader(f)
    for line in reader:
        sentences = json.loads(line['Input.json_data'])['sentences']
        annotations = []
        for x in json.loads(line['Answer.answer_spans']):
            if x['notMention']:
                continue
            querySpan = x['querySpan']
            if x['notPresent']:
                antecedent = []
            else:
                antecedent = x['span_list']
            annotations.append({
                "query": querySpan,
                "antecedents": antecedent
            })
        output["se01e02s0" + str(j)] = {
            "sentences": sentences,
            "annotations": annotations
        }
        j += 1

        # output.append({
        #     "sentences": sentences,
        #     "annotations": annotations
        # })
```
