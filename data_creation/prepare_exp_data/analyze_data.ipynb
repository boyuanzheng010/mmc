```python
import jsonlines
import json
import random
```

```python
def fetch_ratios(temp):
    ratio_list = []
    for i, sample in enumerate(temp):
        doc = []
        for sent in sample['sentences']:
            doc.extend(sent)
        mentions = []
        for cluster in sample['clusters']:
            mentions.extend(cluster)
        ratio_list.append(len(mentions)/len(doc)*100)
    return ratio_list
```

```python
for split in ["train", "dev", "test"]:
    # if split != "dev":
    #     continue
    print(split)
    temp = []
    with open('data/experiment_inputs/dialogue_english/'+split+".english.512.jsonlines") as f:
        reader = jsonlines.Reader(f)
        for line in reader:
            temp.append(line)

    print(fetch_ratios(temp))
```
