## Generate Dropped Version of Dataset

```python
# OntoNotes
drop_ratio = 0.7
for split in ["train", "dev", "test"]:
    if split != "train":
        continue
    print(split)
    temp = []
    with open('data/experiment_inputs/ontonotes_english/'+split+".english.512.jsonlines") as f:
        reader = jsonlines.Reader(f)
        for i, line in enumerate(reader):
            new_clusters = generate_dropped_clusters_ontonotes(line['clusters'], drop_ratio)
            if not new_clusters:
                continue
            line['clusters'] = new_clusters
            temp.append(json.dumps(line))
    with open('data/partial_experiment_inputs/on_en_drop_'+str(int(10*drop_ratio))+"/"+split+".english.512.jsonlines", 'w') as output_file:
        for document in temp:
            output_file.write(document)
            output_file.write('\n')
```

```python
# MMC
drop_ratio = 0.7
for split in ["train", "dev", "test"]:
    if split != "train":
        continue
    print(split)
    temp = []
    with open('data/experiment_inputs/dialogue_english/'+split+".english.512.jsonlines") as f:
        reader = jsonlines.Reader(f)
        for i, line in enumerate(reader):
            new_clusters = generate_dropped_clusters_mmc(line['clusters'], drop_ratio)
            if not new_clusters:
                continue
            line['clusters'] = new_clusters
            temp.append(json.dumps(line))
    with open('data/partial_experiment_inputs/mmc_en_drop_'+str(int(10*drop_ratio))+"/"+split+".english.512.jsonlines", 'w') as output_file:
        for document in temp:
            output_file.write(document)
            output_file.write('\n')
```
