## Build Male and Female List

```python
male_appendix = ["shelldon", "elon", "petrie", "jill", "howie", "rajj", "shedon", "dimitri", "chandler", "kripke", "gablehauser", "mr", "male", "hernandez", "raj", "roeger", "gablehouser", "", ]
female_appendix = ["franny", "ronni", "gloria", "kiki", "priya", "lakshmi", "katee", "bermadette", "missy", "emly", "mrs", "ms", "female", "mrs.", "", "", "", "", "", "", ]

male_names = []
with open('data/raw_source/name_dict/male_top_300.txt', 'r', encoding='utf-8') as f:
    for x in f:
        temp_string = x.strip()
        new_string = temp_string.lower()
        # new_string = temp_string[0] + temp_string[1:].lower()
        male_names.append(new_string)
male_names.extend(male_appendix)

female_names = []
with open('data/raw_source/name_dict/female_top_300.txt', 'r', encoding='utf-8') as f:
    for x in f:
        temp_string = x.strip()
        new_string = temp_string.lower()
        # new_string = temp_string[0] + temp_string[1:].lower()
        female_names.append(new_string)
female_names.extend(female_appendix)

all_names = {
    "male_names": male_names,
    "female_names": female_names
}
import pickle as pkl
with open('name_list.pkl', 'wb') as f:
    pkl.dump(all_names, f)
```

```python
with open('name_list.pkl', 'rb') as f:
    all_names = pkl.load(f)
    intersection = set(all_names['male_names']) & set(all_names['female_names'])
    male_names = []
    for item in all_names['male_names']:
        if item not in intersection:
            male_names.append(item)
    female_names = []
    for item in all_names['female_names']:
        if item not in intersection:
            female_names.append(item)
```

## Modify Names in input files

```python
from utils.convert_speaker_names import build_speaker_dict

count_total = 0
count_modified = 0

for partition in ["dev", "test", "train"]:
    # if partition != "dev":
    #     continue
    data = []
    with open('data/conll_style/dialogue_english/'+partition+'.english.v4_gold_conll', 'r', encoding='utf-8') as f:
        for x in f:
            data.append(x.strip().split())


    # Split into scenes
    scenes = []
    temp = []
    for x in data:
        # Locate the start
        if len(x)==5:
            temp = []
        temp.append(x)
        if len(x)==2:
            scenes.append(temp)

    # Modify Scene Lines
    output = []
    for i, scene in enumerate(scenes):
        # if i>2:
        #     continue
        # Get Speaker Name Map
        speaker_dict = build_speaker_dict(scene, all_names, male_names, female_names)

        # Modify Scene
        for line in scene:
            flag = False
            if len(line)==13:
                if line[8].lower() in speaker_dict:
                    line[8] = speaker_dict[line[8].lower()]
                    flag = True
            elif len(line)>13:
                if line[9].lower() in speaker_dict:
                    line[9: len(line)-4] = [speaker_dict[line[9].lower()]]
                    flag = True
                if line[3].lower() in speaker_dict:
                    line[3] = speaker_dict[line[3].lower()]
                    flag = True
            output.append(" ".join(line)+"\n")
            if len(line)>=13:
                count_total += 1
                if flag:
                    count_modified += 1

    with open('data/conll_style/dialogue_name_replaced_3_english/'+partition+'.english.v4_gold_conll', 'w', encoding='utf-8') as f:
        f.writelines(output)

    print(count_modified, count_total, count_modified/count_total*100)
```

```python
from copy import deepcopy
import random

lengths = set()
# Analyze all speaker names
all_speakers = set()
for i, scene in enumerate(scenes):
    scene_speakers = set()
    # if i!=0:
    #     continue

    # Collect Speaker Names
    for line in scene:
        lengths.add(len(line))
        if len(line)==13:
            scene_speakers.add(line[8].lower())
            all_speakers.add(line[8].lower())
        elif len(line)>13:
            scene_speakers.add(line[9].lower())
            all_speakers.add(line[9].lower())
            # all_speakers.add(" ".join([token.lower() for token in line[9:len(line)-4]]))
    # Build Speaker Dict
    speaker_dict = {}
    male_subset = deepcopy(all_names['male_names'][:100])
    female_subset = deepcopy(all_names['female_names'][:100])
    # print(male_subset)
    # print(female_subset)
    for speaker in scene_speakers:
        if speaker in male_names:
            mapped_name = random.sample(male_subset, 1)[0]
            male_subset.remove(mapped_name)
            speaker_dict[speaker] = mapped_name.capitalize()
        elif speaker in female_names:
            # print("Female:", speaker)
            mapped_name = random.sample(female_subset, 1)[0]
            female_subset.remove(mapped_name)
            speaker_dict[speaker] = mapped_name.capitalize()
    print(speaker_dict)
    print("=="*50)
```
