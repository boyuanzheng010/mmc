```python
input_root = "parallel_data/"
output_root = "scenes/"

for file_name in os.listdir(input_root):
    # Decide which transcript is belongs
    transcript = file_name.strip().split('.')[0].strip().split("_")[1]
    if transcript=="tbbt":
        label = "t"
    elif transcript=="friends":
        label = "f"
    else:
        label = "fe"

    # Extract Scenes for annotation file
    with open(input_root+file_name, 'rb') as f:
        input_data = pkl.load(f)
        scenes = extract_scenes(input_data, label)
        with open(output_root+file_name, 'wb') as f:
            pkl.dump(scenes, f)

    print(file_name, len(input_data), len(scenes))
```

```python
# Collect train
train = []
for i, file_name in enumerate(sorted(os.listdir("scenes/"), reverse=True)):
    if file_name.strip().split('_')[0]=="train":
        with open('scenes/'+file_name, 'rb') as f:
            train.extend(pkl.load(f))
            print(file_name.strip().split('_')[0], len(train))
print()

dev = []
for i, file_name in enumerate(sorted(os.listdir("scenes/"), reverse=True)):
    if file_name.strip().split('_')[0]=="dev":
        with open('scenes/'+file_name, 'rb') as f:
            dev.extend(pkl.load(f))
            print(file_name.strip().split('_')[0], len(dev))
print()

test = []
for i, file_name in enumerate(sorted(os.listdir("scenes/"), reverse=True)):
    if file_name.strip().split('_')[0]=="test":
        with open('scenes/'+file_name, 'rb') as f:
            test.extend(pkl.load(f))
            print(file_name.strip().split('_')[0], len(test))

with open('scenes/train_all.pkl', 'wb') as f:
    pkl.dump(train, f)
with open('scenes/dev_all.pkl', 'wb') as f:
    pkl.dump(dev, f)
with open('scenes/test_all.pkl', 'wb') as f:
    pkl.dump(test, f)
```
