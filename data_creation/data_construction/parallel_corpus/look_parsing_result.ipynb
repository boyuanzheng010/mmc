```python
import pickle as pkl
import spacy
```

```python
sm_parser = spacy.load('en_core_web_sm')
```

```python
with open('parallel_data/parsed_corpus_all_zh.pkl', 'rb') as f:
    data = pkl.load(f)
```

```python
print(data.keys())
```

```python
def merge_maximum_span(spans):
    spans.sort(key=lambda x: x[1])
    # print(spans)
    to_pop = []
    for j, (word_0, start_idx_0, end_idx_0) in enumerate(spans):
        for k, (word_1, start_idx_1, end_idx_1) in enumerate(spans):
            if k==j:
                continue
            if (start_idx_1 >= start_idx_0) and (end_idx_1 <= end_idx_0):
                to_pop.append(spans[k])
    for item in to_pop:
        if item in spans:
            spans.remove(item)

    spans.sort(key=lambda x: x[1])

    return spans
```

```python
for scene_key in data:
    for scene in data[scene_key]:
        for utt in scene:
            noun_phrase = merge_maximum_span(list(set(utt['sm_noun_chunk']) | set(utt['berkeley_noun_chunk']) | set(utt['trf_noun_chunk'])))
            for token in noun_phrase:
                if token[0].count("\"")%2!=0:
                    sentence_token = [item.text for item in sm_parser(utt["sentence"])]
                    print(token[0].count("\""), token[0].count("\"")%2)
                    print(utt["sentence"])
                    print(token)
                    print(sentence_token[token[1]: token[2]])
                    if sentence_token[token[2]]=="\"":
                        print(sentence_token[token[1]: token[2]+1])
                    print("=="*50)
```
