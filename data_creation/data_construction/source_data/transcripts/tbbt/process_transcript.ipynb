# Process Transcript Corpus

1.Collect Transcript
The file data is "original_corpus.json" file.
This file is collected by crawling transcript in open source website.
The code base for scrapper and data processing locates at server_name_2 Grid: /home/user_id_1/multi_coref/resources/TBBTCorpus

2.Parser each transcript utterance into speaker, sentence.
We fetch the speaker by splitting the sentence with the first ":".

3.Perform text cleaning in speaker, sentence (including brackets, and some appendix information)
We clean brackets in speaker.



*Extra:
We further clean setence and keep the original sentence since it contains extra information. This may be useful for the follow-up research. This text is store in the "sentence_with_infor" field in each utterance instance.

```python
import csv
import pickle as pkl
import json
import re
```

```python
def clean_sentence_brackets(text):
    text = text.strip().lstrip('-').lstrip().lstrip('.').lstrip()
    new_text = re.sub(u"\\(.*?\\)|\\{.*?}|\\[.*?]", "", text)
    return new_text
```

```python
# Load Corpus
with open('original_corpus.json',"r") as corpus_file:
	corpus = json.load(corpus_file)
	corpus_file.close()
```

```python
data = {}
for epi_key in corpus:
    output = []
    for x in corpus[epi_key].strip().split("\n"):
        if ":" not in x:
            continue
        temp = x.strip().split(":")
        speaker = clean_sentence_brackets(temp[0])
        utterance = ":".join(temp[1:]).strip().lstrip('-').lstrip().lstrip('.').lstrip().strip(":").strip()
        output.append([utterance, speaker])
    temp_key = tuple([int(item) for item in epi_key.strip().split("_")])
    data[temp_key] = output[:-1]
```

```python
with open('tbbt_transcripts.pkl', 'wb') as f:
    pkl.dump(data, f)
```
